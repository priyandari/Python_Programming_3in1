## JSON (JavaScript Object Notation)

File __`*.json (JavaScript Object Notation)`__ adalah file yang menyimpan data berbentuk _multi dictionary_. File *.json dapat menyimpan data yang beragam secara terstruktur. 

Format struktur data json sering digunakan sebagai format untuk __pertukaran data__ antar komputer atau antar program, yang diberikan authentikasi.

Format data json lebih mudah di parsing oleh python ke dalam list dan dictionary.


Contoh septong data json
```python
{
    "id": 237478,
    "city": {
        "id": {
            "$numberLong": "237478"
        },
        "name": "Mobaye",
        "findname": "MOBAYE",
        "country": "CF",
        "coord": {
            "lon": 21.17861,
            "lat": 4.31902
        },
        "zoom": {
            "$numberLong": "10"
        }
    }
}
```

Mengakses file json membutuhkan library json. Library ini umumnya sudah bawaan Python. Jika belum ada bisa ditambahkan install.

````python
# Mengakses file json
import json

DATA_FILE_NAME = 'files/history_city_list.json'
with open(DATA_FILE_NAME, 'r', encoding='Latin1') as jsonfile:
    data = json.load(jsonfile)

print(type(data))
print(len(data))

# print(data[764])
print(json.dumps(data[764], indent=4))
````
